```python
import numpy as np
import time
from torch.autograd import grad, Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(12345)
lims=(-5, 5)

use_cuda = torch.cuda.is_available()
use_cuda, torch.version.cuda, torch.__version__
```

# Define sampler from real data and Z 

```python
from scipy.stats import rv_discrete

MEANS = np.array(
        [[-1,-3],
         [1,3],
         [-2,0],
        ])
COVS = np.array(
        [[[1,0.8],[0.8,1]],
        [[1,-0.5],[-0.5,1]],
        [[1,0],[0,1]],
        ])
PROBS = np.array([
        0.2,
        0.5,
        0.3
        ])
assert len(MEANS) == len(COVS) == len(PROBS), "number of components mismatch"
COMPONENTS = len(MEANS)

comps_dist = rv_discrete(values=(range(COMPONENTS), PROBS))

def sample_true(N):
    comps = comps_dist.rvs(size=N)
    conds = np.arange(COMPONENTS)[:,None] == comps[None,:]
    arr = np.array([np.random.multivariate_normal(MEANS[c], COVS[c], size=N)
                     for c in range(COMPONENTS)])
    return np.select(conds[:,:,None], arr).astype(np.float32)

NOISE_DIM = 20
def sample_noise(N):
    return np.random.normal(size=(N,NOISE_DIM)).astype(np.float32)
```

# Define architectures

After you've passed task 1 you can play with architectures.

#### Generator

```python
class Generator(nn.Module):
    def __init__(self, noise_dim, out_dim, hidden_dim=100):
        super(Generator, self).__init__()
        
        self.fc1 = nn.Linear(noise_dim, hidden_dim)
        nn.init.xavier_normal(self.fc1.weight)
        nn.init.constant(self.fc1.bias, 0.0)
        
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc2.weight)
        nn.init.constant(self.fc2.bias, 0.0)
        
        self.fc3 = nn.Linear(hidden_dim, out_dim)
        nn.init.xavier_normal(self.fc3.weight)
        nn.init.constant(self.fc3.bias, 0.0)

    def forward(self, z):
        """
            Generator takes a vector of noise and produces sample
        """
        h1 = F.tanh(self.fc1(z))
        h2 = F.leaky_relu(self.fc2(h1))
        y_gen = self.fc3(h2)
        return y_gen
```

#### Discriminator

```python
class Discriminator(nn.Module):
    def __init__(self, in_dim, hidden_dim=100):
        super(Discriminator, self).__init__()
        
        self.fc1 = nn.Linear(in_dim, hidden_dim)
        nn.init.xavier_normal(self.fc1.weight)
        nn.init.constant(self.fc1.bias, 0.0)
        
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc2.weight)
        nn.init.constant(self.fc2.bias, 0.0)
        
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        nn.init.xavier_normal(self.fc3.weight)
        nn.init.constant(self.fc3.bias, 0.0)
        
        self.fc4 = nn.Linear(hidden_dim, 1)
        nn.init.xavier_normal(self.fc4.weight)
        nn.init.constant(self.fc4.bias, 0.0)

    def forward(self, x):
        h1 = F.tanh(self.fc1(x))
        h2 = F.leaky_relu(self.fc2(h1))
        h3 = F.leaky_relu(self.fc3(h2))
        score = self.fc4(h3)
        if TASK == 1 or TASK == 2:
            score = F.sigmoid(self.fc4(h3))
        return score
```

# Define updates and losses

```python
generator = Generator(NOISE_DIM, out_dim = 2)  
discriminator = Discriminator(in_dim = 2)

if use_cuda:
    generator = generator.cuda()
    discriminator = discriminator.cuda()


lr = 0.001

g_optimizer = optim.Adam(generator.parameters(),     lr=lr, betas=(0.5, 0.999))
d_optimizer = optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))
```

Notice we are using ADAM optimizer with `beta1=0.5` for both discriminator and discriminator. This is a common practice and works well. Motivation: models should be flexible and adapt itself rapidly to the distributions. 

You can try different optimizers and parameters.

```python
################################
# IMPLEMENT HERE
# Define the g_loss and d_loss here
# these are the only lines of code you need to change to implement GAN game
EPS = 1e-6
def g_loss(answer_gen):
    if TASK == 1: 
        return torch.mean(torch.log(1 - answer_gen + EPS))
    if TASK == 2:
        return -torch.mean(torch.log(answer_gen + EPS))
    if TASK == 3 or TASK == 4:
        return -torch.mean(answer_gen)
    raise ValueError()
    
def d_loss(answer_gen, answer_inp):
    if TASK == 1 or TASK == 2:
        return -torch.mean(torch.log(answer_inp + EPS)) - torch.mean(torch.log(1 - answer_gen + EPS))
    if TASK == 3 or TASK == 4:
        return torch.mean(answer_gen) - torch.mean(answer_inp)
    raise ValueError()

################################
```

# Get real data

```python
data = sample_true(100000)
def iterate_minibatches(X, batchsize, y=None):
    perm = np.random.permutation(X.shape[0])
    
    for start in range(0, X.shape[0], batchsize):
        end = min(start + batchsize, X.shape[0])
        if y is None:
            yield X[perm[start:end]]
        else:
            yield X[perm[start:end]], y[perm[start:end]]
```

# Train the model

```python
def calc_gradient_penalty(discriminator, real_data, fake_data, lambda_=.1):
    alpha = torch.rand(real_data.shape[0], 1).expand(real_data.size())
    alpha = alpha.cuda() if use_cuda else alpha

    interpolates = alpha*real_data + ((1 - alpha)*fake_data)

    if use_cuda:
        interpolates = interpolates.cuda()
    interpolates = Variable(interpolates, requires_grad=True)

    disc_interpolates = discriminator(interpolates)
    grad_outputs = torch.ones(disc_interpolates.size())
    grad_outputs = grad_outputs.cuda() if use_cuda else grad_outputs
    gradients = grad(outputs=disc_interpolates, inputs=interpolates, grad_outputs=grad_outputs,
                     create_graph=True, retain_graph=True, only_inputs=True)[0]
    
    return ((gradients.norm(2, dim=1) - 1)**2).mean()*lambda_
```

```python
from IPython import display

def train(num_epochs=20, batch_size=64):
    plt.xlim(lims)
    plt.ylim(lims)

    # ===========================
    # IMPORTANT PARAMETER:
    # Number of D updates per G update
    # ===========================
    k_d, k_g = 4, 1
    if TASK == 3 or TASK == 4:
        k_d = 5

    accs = []
    n_batches = len(data)//batch_size + 1

    try:
        for epoch in range(num_epochs):
            for i, input_data in enumerate(iterate_minibatches(data, batch_size)):
                d_loss_val = 0
                g_loss_val = 0

                # Optimize D
                for _ in range(k_d):
                    # Sample noise
                    noise = Variable(torch.Tensor(sample_noise(len(input_data))))

                    # Do an update
                    inp_data = Variable(torch.Tensor(input_data))
                    if use_cuda:
                        noise = noise.cuda()
                        inp_data = inp_data.cuda()
                    data_gen = generator(noise)
                    loss = d_loss(discriminator(data_gen), discriminator(inp_data))
                    if TASK == 4:
                        loss += calc_gradient_penalty(discriminator, inp_data.data, data_gen.data, batch_size)
                    d_optimizer.zero_grad()
                    loss.backward()
                    d_optimizer.step()

                    d_loss_val = loss.data.cpu().numpy()[0]
                
                    if TASK == 3:
                        for w in discriminator.parameters():
                            w.clamp(-0.01, 0.01)
                    
                    
                # Optimize G
                for _ in range(k_g):
                    # Sample noise
                    noise = Variable(torch.Tensor(sample_noise(len(input_data))))

                    # Do an update
                    if use_cuda:
                        noise = noise.cuda()
                    data_gen = generator(noise)
                    loss = g_loss(discriminator(data_gen))
                    g_optimizer.zero_grad()
                    loss.backward()
                    g_optimizer.step()

                    g_loss_val = loss.data.cpu().numpy()[0]

                if i % 100 == 0:
                    # Visualize
                    display.clear_output(wait=True)
                    plt.clf()
                    vis_data(data); vis_g(); vis_d()
                    plt.show()

                    print(f'\r epoch #{epoch + 1}/{num_epochs}, batch #{i + 1}/{n_batches} d_loss={d_loss_val:.4f}, g_loss={g_loss_val:.4f}')


    except KeyboardInterrupt:
        pass
```
